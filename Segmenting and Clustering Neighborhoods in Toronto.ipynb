```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
```

```python
dfs = pd.read_html(url, header=0)
```

The table we would like to scrape for data is the first table of the webpage.

```python
table_df = dfs[0]
```

Pandas has read the .html into a table with three (3) columns, which is precisely the number of columns we want.
The column names are also correct and in the desired order.

```python
table_df.info()
```

```python
table_df
```

```python
table_df_with_NaN = table_df.replace(to_replace='Not assigned', value=np.nan, inplace=False)
table_df_with_NaN
```

```python
table_df_dropped_NaN = table_df_with_NaN.dropna(axis='index', how='any', inplace=False)
table_df_dropped_NaN
```

```python
# This line returns the number of unique postal codes in the DataFrame.
print('Number of unique postal codes in table: {}'.format(len(pd.unique(table_df_dropped_NaN['Postal Code']))))
```

```python
postal_codes_df = table_df_dropped_NaN.sort_values(by='Postal Code', axis='index', inplace=False).reset_index(inplace=False).drop(labels='index', axis='columns', inplace=False)
postal_codes_df
```
